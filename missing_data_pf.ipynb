## The Why

Missing data is a challenge faced by localization filters. An autonomous vehicle driving in snowy conditions, for example, loses the ability to identify its location based upon the surface of the road expected to be seen by the vehicle.

## Proposed Solutions

Many solutions to the more general Bayesian Missing Data Problem have been proposed. These include Expectation Maximization, Data Augmentation, and Noniterative Computation (Ming T. Tan, Guo-Liang Tian, Kai Wang Ng, 2010). For example, data augmentation can combine GPS location data with data sensed locally around the vehicle. In the particle filter localization sphere, Multiple Imputations Particle Filters (MIPF) have also been suggested. These use randomly drawn imputation values which can converge in non-linear systems (Xiao-Ping Zhang, Ahmed Shaharyar Khwaja, Ji-An Luo, and Alagan Anpalagan, 2015).

## Experiment

This experiment explores the missing data problem for a robotic kit car. The code provided here is self contained and runs without any real-world hardware sensors. But the code can be adapted to run similar experiments on robotic car hardware.

![Particle Filter Localization](images/particle-filter-localization-004.png)

## Sensor Data

The experiment runs in a 1-dimensional environment with a robotic car driving slow and straight towards a facing wall. It is assumed the robotic car has access to three different sensors:

1. **Ultrasonic Distance Sensor:** The ultrasonic sensor is used by the car to detect the distance to the wall ahead.


2. **Line Detector:**. In this experiment, the robotic car's line detector is not used in the usual way to guide the robot along a pre-defined path. Instead, the line detector is used to detect painted lines (or black tape stuck to the floor) striped perpendicular to the car's path. The distance between each stripe is known to the car's internal model (or map) of the world.


3. **Tilt-Pan Camera:** The rotating camera is used to detect the angle to one or more landmarks which run alongside the path of the car.

![Robot Car Sensors](images/robot-sensors-002.png)

## Missing Data

There are several causes of missing data. First, the robotic car may be too far away from the facing wall to measure its distance. Second, 'snow' may be covering the striped lines (this would be simulated by simply removing the black tape from the floor). And third, there will be periods of time in which the landmarks would be out of visibility as the camera sweeps back-and-forth across the car.

## The Code

```python
import math
import random
import matplotlib.pyplot as plt
```

## Landmarks

There are three different locators that the vehicle sensor can pick up:
- Landmarks - located each meter at 1 meter distance from the path
- Obstacles - the single wall located at the origin
- Stripes - located each meter having a thickness of 10 cm

```python
# The vehicle drives slow and straight along a 1-dimensional line at y_position == 0.0

# All of the landmarks have a y_position of 1.0 meters away from the straight-line path of the vehicle
landmarks = [[-4.0, 1.0], # x_position, y_position
             [-3.0, 1.0],
             [-2.0, 1.0],
             [-1.0, 1.0],
             [ 0.0, 1.0]
             ]

# The facing wall is the only obstacle
obstacles = [[0.0, 0.0]] # x_position, y_position

# A stripe is placed each meter across the vehicle's path with each stripe
WIDTH_OF_STRIPE = 0.10 # 10 cm wide
stripes = [[-5.0, -5.0 + WIDTH_OF_STRIPE], # x_position_start, x_position_end in meters
           [-4.0, -4.0 + WIDTH_OF_STRIPE],
           [-3.0, -3.0 + WIDTH_OF_STRIPE],
           [-2.0, -2.0 + WIDTH_OF_STRIPE],
           [-1.0, -1.0 + WIDTH_OF_STRIPE],
           ]

# Global Constants
STARTING_WORLD_X_POSITION = -5.0 # The starting location of the 1-dimensional world along the x-axis
ENDING_WORLD_X_POSITION = 0.0 # The ending location of the 1-dimensional world along the x-axis
NUMBER_OF_TRACKING_PARTICLES = 1000 # The number of particles tracking the location of the robot vehicle
NUMBER_OF_SIMULATION_ITERATIONS = 100 # The number of iterations in the simulation

STARTING_VEHICLE_X_POSITION = -4.5 # The starting x_location of the robot vehicle
STARTING_VEHICLE_X_VELOCITY = abs(STARTING_VEHICLE_X_POSITION / NUMBER_OF_SIMULATION_ITERATIONS) # The constant velocity of the robot vehicle and all particles
STARTING_VEHICLE_Y_POSITION = 0.0 # The constant y_position of the robot vehicle and all particles
STARTING_VEHICLE_TIME_DELTA = 1.0 # The time delta (dt) in seconds by which the simulation updates

PARTICLE_X_VELOCITY_NOISE = 0.01 # meters/second (standard deviation) - the amount a particle can drift when moved to a new x_position
PARTICLE_X_POSITION_NOISE = 0.05 # meters along the path (standard deviation) - the x_position drift when detecting path stripe locations along the path and distance to the facing wall
PARTICLE_Y_POSITION_NOISE = 0.10 # meters from wall (standard deviation) - the y_position off the path when measuring angle to the landmark

GUASSIAN_ANGLE_VARIANCE = math.pi / 4.0 # The variance used by the gaussian_probability() function when determining the probability of whether two angles match
GUASSIAN_DISTANCE_VARIANCE = 2.0 # The variance used by the gaussian_probability() function when determining the probability of whether two distances match

STRIPE_HIT_PROBABILITY_MULTIPLE = (1.0 - WIDTH_OF_STRIPE) / len(stripes) # If both the robot and the particle see a stripe then there is a 98% chance / number of stripes they share the same location
STRIPE_MISS_PROBABILITY_MULTIPLE = 1.0 - STRIPE_HIT_PROBABILITY_MULTIPLE # If the robot sees a stripe but the particle does not then don't completely kill the probability weight of the particle

LANDMARKS_VISIBILITY = 0.1 # The probability that a landmark is visible at any given iteration
OBSTACLES_VISIBLITY = 0.1 # The probability that an obstacle is visible at any given iteration
STRIPES_VISIBLITY = 0.2 # The probability that a stripe is visible at any given iteration
```

## Particles

1,000 particles are initially spread uniformly along the vehicle's 1-dimensional path. Each particle can make a noisy measurement to each of the sensor locations. The robot is also a particle, but unlike the particles, the robot sensor data may be blocked by poor visibility.

```python
# Global Variables
particle_id_counter = 0 # Automatically increment the particle ID counter whenever a new particle is created
active_particles = [] # List of active particles tracking the robot vehicle - this list will be probabilistically updated each simulation iteration

class Particle(object):
    """
    Only the Horizontal Position (x) is changing with Angle (theta) depending upon (x,y) position.
    Assumes the Vertical Position (y) and Velocity are constant, although
    Vertical Postition and Velocity experiences minor drift variation with each call.
    Assumes random process noise is also added to the measured angle.

    Note that the robot itself is also instantiated as a 'Particle'.
    Both the robot's and the particle's measurements and motions will suffer from noise.
    But the particle will always be able to see all landmarks, stripes, and obstacles.
    The robot will only be able to see landmarks, stripes, and obstacles if they have
    not been somehow obscured by snow, trees, or camera angle.
    """

    def __init__(self, x_position=-5.0, x_velocity=0.1, y_position=0.0, dt=1.0):
        """
        We want to localize the position of the vehicle using particles assuming
        a constant velocity and fixed y_position  of the landmark relative to the path.
        Our state variables are: x_position, x_velocity, and y_position.
        """
        global particle_id_counter
        self.particle_id = particle_id_counter # The unique particle id assigned to this particle (id == 0 indicates this is the robot itself)
        self.original_id = particle_id_counter # The original particle id if this particle was cloned
        particle_id_counter += 1 # Automatically increment the particle id everytime a new Particle is created
        self.x_position = x_position # meters located along the track
        self.x_velocity = x_velocity # meters/second along track (constant other than drift noise)
        self.y_position = y_position # meters from wall (constant although measurements can drift from this position)
        self.dt = dt # seconds (constant)
        self.weight = 0.0 # The non-normalized probability weight that the location of this particle matches the location of the robot

    def __str__(self):
        return f'[id={self.particle_id}, x={self.x_position:.3f}, v={self.x_velocity:.3f}, y={self.y_position:.3f}, t={self.dt:.3f}]'

    def __repr__(self):
        return f'id,{self.particle_id},x,{self.x_position:.3f},v,{self.x_velocity:.3f},y,{self.y_position:.3f},t,{self.dt:.3f},'

    def is_robot(self):
        """
        This Particle is the robot itself if it has a particle_id of zero.
        """
        return self.particle_id == 0 # The robot must be the first particle to be instantiated

    def clone(self):
        """
        Clone this Particle when sampling with replacement.
        The new particle will be located at a fuzzy distance from the old particle.
        """
        x_position = self.x_position + random.gauss(0.0, PARTICLE_X_POSITION_NOISE) # Clone the new particle at a fuzzy distance from the old particle
        particle = Particle(x_position=x_position, x_velocity=self.x_velocity, y_position=self.y_position, dt=self.dt)
        particle.original_id = self.particle_id
        return particle

    def move(self):
        """
        Move the particle at a pre-defined fixed velocity for a fixed time.
        While the vehicle can only move in a straight line, the velocity of the particles are subject to random noise.
        This will fuzz the location of identical particles.
        """
        velocity = self.x_velocity + random.gauss(0.0, PARTICLE_X_VELOCITY_NOISE)
        self.x_position += velocity * self.dt

    def sense_landmarks(self, visibility_mask=None):
        """
        Use the tilt-pan camera to measure the angle to the landmarks.
        Guassian noise in the y_position of the particle will cause the actual distance to vary from the returned measured distance.
        If a visibility_mask has been passed as an array then True indicates the landmark is visible.
        """
        angles=[]
        for landmark_id in range(len(landmarks)):
            landmark = landmarks[landmark_id]
            if visibility_mask != None and not visibility_mask[landmark_id]:
                angles.append(None)
                continue
            delta_x = landmark[0] - self.x_position + random.gauss(0.0, PARTICLE_X_POSITION_NOISE)
            delta_y = landmark[1] - self.y_position + random.gauss(0.0, PARTICLE_Y_POSITION_NOISE)
            angle = math.atan2( delta_y, delta_x )
            angles.append(angle)
        return angles

    def sense_obstacles(self, visibility_mask=None):
        """
        Use the ultrasound distance sensor to measure the distance to the obstacles.
        Guassian noise in the x_position and y_position of the particle will cause the actual distance to vary from the returned measured distance.
        If a visibility_mask has been passed as an array then True indicates the obstacle is visible.
        """
        distances=[]
        for obstacle_id in range(len(obstacles)):
            obstacle = obstacles[obstacle_id]
            if visibility_mask != None and not visibility_mask[obstacle_id]:
                distances.append(None)
                continue
            delta_x = obstacle[0] - self.x_position + random.gauss(0.0, PARTICLE_X_POSITION_NOISE)
            delta_y = obstacle[1] - self.y_position + random.gauss(0.0, PARTICLE_Y_POSITION_NOISE)
            distance = math.sqrt( delta_x**2.0 + delta_y**2.0 )
            distances.append(distance)
        return distances

    def sense_stripes(self, visibility_mask=None):
        """
        Use the line sensor to detect whether the vehicle is positioned above a stripe.
        Guassian noise in the x_position of the particle will cause the sensor result to vary from the actual position.
        If a visibility_mask has been passed as an array then True indicates the stripe is visible.
        Only a single True (stripe detected) or False (no stripe detected) will be returned.
        """
        detected = False
        for stripe_id in range(len(stripes)):
            stripe = stripes[stripe_id]
            if visibility_mask != None and not visibility_mask[stripe_id]:
                continue
            measured_x = self.x_position + random.gauss(0.0, PARTICLE_X_POSITION_NOISE)
            if measured_x >= stripe[0] and measured_x <= stripe[1]:
                detected = True
                break
        return detected

    def gaussian_probability(self, robot_measurement, particle_measurement, sigma):
        """
        The Guassian Probability function calculate the probability that the sample
        falls within the 1-dimensional Guassian distribution having mean mu and variance sigma.
        For the particle filter, the Guassian Probability function calculates the probability
        that particle shares the same location as the robot from a comparison of their measurements.
        """
        return (1.0 / math.sqrt( 2.0 * math.pi * (sigma**2.0))) * math.exp(- ((particle_measurement - robot_measurement)**2.0 ) / (sigma**2.0) / 2.0)

    def calculate_matching_location_probability_weight(self, robot_landmarks, robot_obstacles, robot_stripes):
        """
        Calculates the cumulative, non-normalized probability weight that this particle shares the same
        location as the robot based on a comparison of all their measurements.
        """
        probability = 1.0 # Starting uniform probability

        # Check the landmark measurements
        if robot_landmarks != None: # The robot may not be able to sense landmarks
            particle_landmarks = self.sense_landmarks() # But the particle can always sense all landmarks
            assert len(robot_landmarks) == len(particle_landmarks)
            for landmark_id in range(len(robot_landmarks)):
                robot_landmark = robot_landmarks[landmark_id]
                particle_landmark = particle_landmarks[landmark_id]
                if robot_landmark != None and particle_landmark != None:
                    probability *= self.gaussian_probability(robot_landmark, particle_landmark, GUASSIAN_ANGLE_VARIANCE) # Using angle variance to calculate probability

        # Check the obstacle measurements
        if robot_obstacles != None: # The robot may not be able to sense obstacles
            particle_obstacles = self.sense_obstacles() # But the particle can always sense all obstacles
            assert len(robot_obstacles) == len(particle_obstacles)
            for obstacle_id in range(len(robot_obstacles)):
                robot_obstacle = robot_obstacles[obstacle_id]
                particle_obstacle = particle_obstacles[obstacle_id]
                if robot_obstacle != None and particle_obstacle != None:
                    probability *= self.gaussian_probability(robot_obstacle, particle_obstacle, GUASSIAN_DISTANCE_VARIANCE) # Using distance variance to calculate probability

        # Check the stripe measurements
        if robot_stripes == True: # The fact that no stripe is seen by the robot provides no information as stripes are small and it could be snowing
            particle_stripes = self.sense_stripes() # The particle can always see the stripes
            if particle_stripes:
                probability *= STRIPE_HIT_PROBABILITY_MULTIPLE # If both the robot and the particle see a stripe then there is a high chance they share the same location
            else:
                probability *= STRIPE_MISS_PROBABILITY_MULTIPLE # If the robot sees the strip but the particle doesn't then there is a low chance they share the same location

        # Set the non-normalized probability weight of this particle
        self.weight = probability
```

## Manage Particles

The particles move at the same velocity as the robot. Particles that have a high probability of being located at the same or near the robot are replicated each iteration. Each iteration calculates the cumulative, non-normalized probability weights of each of the active particles and makes a comparison to the robot's measurements (unless poor visibility makes the robot's measurement unavailable). If no sensors are operational then the non-normalized probability weights == 1.0. This will cause all of the existing particles to be uniformly sampled for the next calculation.

```python
def generate_particles():
    """
    Initializes the set of active particles tracking the robot vehicle.
    This list of active particles will get updated via probabilistic resampling
    each iteration of the simulation.
    """
    active_particles = []
    for id in range(NUMBER_OF_TRACKING_PARTICLES):
        x_location = random.uniform(STARTING_WORLD_X_POSITION, ENDING_WORLD_X_POSITION)
        particle = Particle(x_location, STARTING_VEHICLE_X_VELOCITY, STARTING_VEHICLE_Y_POSITION, STARTING_VEHICLE_TIME_DELTA)
        active_particles.append(particle)
    return active_particles

def move_particles():
    """
    Move each of the active particles according to the particle's pre-defined location and velocity.
    While the vehicle can only move in a straight line, the velocity of the particles are subject to random noise.
    This will fuzz the location of cloned identical particles starting at the same location.
    """
    for particle in active_particles:
        particle.move()
    return active_particles

def update_matching_location_probability_weights(robot_landmarks=None, robot_obstacles=None, robot_stripes=None):
    """
    Calculate the cumulative, non-normalized probability weights of each of the active particles
    in comparison to the robot's measurements (if available).
    If no sensors are operational then the non-normalized probability weights == 1.0.
    This will cause all of the existing particles to be uniformly sampled for the next calculation.
    """
    for particle in active_particles:
        particle.calculate_matching_location_probability_weight(robot_landmarks, robot_obstacles, robot_stripes)
    return active_particles

def get_particle_positions():
    """
    Collect the positions of the latest set of active particles in a list.
    """
    positions = []
    for particle in active_particles:
        position = particle.x_position
        positions.append(position)
    return positions

def get_particle_probability_weights():
    """
    Collect the non-normalized probability weights from each of the active particles.
    """
    particle_weights = []
    for particle in active_particles:
        weight = particle.weight
        particle_weights.append(weight)
    return particle_weights

def resample_particles_by_random_choices(particle_weights):
    """
    Generate a new list of active particles by resampling the previous list of active particles
    and selecting those particles with the highest probability weights (with replacement).
    The new set of particles are cloned versions of the old particles with fuzzy
    """
    global active_particles
    assert len(active_particles) == len(particle_weights)
    selected_particles = random.choices(active_particles, weights=particle_weights, k=len(particle_weights))
    cloned_particles = []
    for particle in selected_particles:
        cloned_particles.append( particle.clone() )
    active_particles = cloned_particles
    return active_particles

def make_visibility_mask(locations, visibility):
    """
    Returns visibility_mask array of True/False boolean values.
    The length of the array depends upon the number of elements in the locations argument.
    The probability that a location is visible is determined by the visibility argument.
    """
    if visibility == None: # Empty visibility probability returns an empty visibility_mask == everything visible
        return None
    visibility_mask=[]
    for location in locations:
        if random.random() <= visibility:
            visibility_mask.append(True)
        else:
            visibility_mask.append(False)
    return visibility_mask
```

## Run the Particle Filter

The motion of the robot is simulated in an iterative fashion. During each iteration, an attempt to take measurements in an environment with poor visibility is made. Then the set of particles tracking the location of the vehicle are evaluated for their probability of matching the vehicle's location. A visibility mask is created based upon the probability that a landmark, obstacle, or stripe is visible at any given iteration. This means that the probabilities that a particle's location matches that of the vehicle cannot be updated due to missing data.

```python
def run_particle_filter_simulation(
        landmarks_visibility_probability=None,
        obstacles_visibility_probability=None,
        stripes_visibility_probability=None,
        output_figure_filename=None):
    """
    Simulate the motion of the robot vehicle.
    Attempt to take measurements in an environment with poor visibility.
    Refine the set of particles tracking the location of the vehicle given substantial missing data.
    landmarks_visibility_probability = The probability that a landmark is visible at any given iteration
    obstacles_visibility_probability = The probability that an obstacle is visible at any given iteration
    stripes_visibility_probability = The probability that a stripe is visible at any given iteration
    """
    # Initialize the simulation
    robot = Particle(STARTING_VEHICLE_X_POSITION, STARTING_VEHICLE_X_VELOCITY, STARTING_VEHICLE_Y_POSITION, STARTING_VEHICLE_TIME_DELTA) # The robot itself will get the particle_id == 0
    global active_particles
    active_particles = generate_particles() # Generate a uniform distribution of particles randomly placed along the 1-dimensional world

    # Iterate the simulation
    for iteration in range(NUMBER_OF_SIMULATION_ITERATIONS):
        
        # Update the visibility of each location
        landmarks_visibility=make_visibility_mask(landmarks, landmarks_visibility_probability)
        obstacles_visibility=make_visibility_mask(obstacles, obstacles_visibility_probability)
        stripes_visibility=make_visibility_mask(stripes, stripes_visibility_probability)

        # Take robot measurements
        robot_landmarks = robot.sense_landmarks(visibility_mask=landmarks_visibility) # Robot measures the angle of all landmarks
        robot_obstacles = robot.sense_obstacles(visibility_mask=obstacles_visibility) # Robot measures the distance to all obstacles
        robot_stripes = robot.sense_stripes(visibility_mask=stripes_visibility) # Robot determines whether it is sitting above a stripe in the path

        # Update the list of active particles by the probability that the particle location matches the robot location
        update_matching_location_probability_weights(robot_landmarks, robot_obstacles, robot_stripes) # Compare the robot's measurements to each particle's measurement
        particle_weights = get_particle_probability_weights() # Collect the probability weights from each of the active particles
        resample_particles_by_random_choices(particle_weights) # Generate a new list of active particles from the most like particles from the previous list

        # Move the robot and the particles
        robot.move()
        move_particles()

    # Collect the positions of the latest set of active particles in a list
    positions = get_particle_positions()

    return positions
```

## Change Visibility

The particle filter can be run both with good visibility (passing no probabilities) or poor visibility (passing the probability that a location is visible).

```python
# Run simulation with perfect visibility
good_visibility_positions = run_particle_filter_simulation()
```

```python
# Run simulation with poor visibility - passing the probability that a location is visible
poor_visibility_positions = run_particle_filter_simulation(LANDMARKS_VISIBILITY, OBSTACLES_VISIBLITY, STRIPES_VISIBLITY)
```

## Charting Results

After 100 iterations, the location of the vehicle should be at the origin (x=0.0). The predictions of the residual particles iteratively selected according to their probability weights can then be compared. The findings below show that the variation in the particle filter results almost doubles when visibility is very poor.
