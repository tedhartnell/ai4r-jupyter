## The Why

Simultaneous Localization and Mapping (SLAM) algorithms typically use Light Detection and Ranging (LiDAR) sensor data along with Inertia Measurement Unit (IMU) sensor data to both localize the position of a robot and map the robot's environment in real-time. Unfortunately, only the most expensive robotic car kits carry such exotic hardware.

SLAM Map is a very simple experimental template for regular robotic car kits. It allows a robot to simultaneously localize its position while it maps a 1-dimensional environment. This code provides a self-contained example that can be run without robotic hardware. But it is designed to be adapted into a real-world experiment as desired.

## Real-World Setup

Run this experiment by placing a robot car kit on a straight, 1-dimensional track. Arrange a 'feature wall' in a line running parallel to the track for the robot to map. The features in the wall should all be approximately the same size as the robot.
Features may include: shoeboxes, tissue boxes, pots, bins, jugs, and other household objects.

Painted lines or black tape should be placed perpendicular to the track at precise 1 meter intervals. The robot will detect these lines as it drives over them using its onboard Line Detector sensor. Detecting these lines will provide the robot with high-confidence landmarks that can be feed into its SLAM algorithm.

At the same time, the robot will be adding low-confidence waypoints to the SLAM matrix at regular intervals of 0.1 second. As the robot has no idea of its speed, it must guess the location of these waypoints between each of the lines. It can do this by guessing that each waypoint is 1 cm from the last waypoint.

For example, the robot might start 10 cm back from the first waypoint. It will then add waypoints to an array until it hits the first painted line. When this happens, the robot will be certain that it is located at the origin point along the track. The robot will then continue adding waypoints (each time guessing that they are located 1 cm apart) while looking out for more painted lines (known to be exactly 1 m apart).

Each time the robot adds a waypoint to the SLAM matrix, it will also record the distance to the feature wall. It takes a distance measurement using an onboard Ultrasonic Sensor pointed perpendicular to the track towards the feature wall.

### Experiment 1
Drive the robot at constant speed while guessing waypoints are separated by 1 cm.

### Experiment 2
Add constant acceleration to the robot while still guessing waypoints are separated by 1 cm.

![SLAM Map Configuration](images/slam-map-configuration-002.png)

## Hardware Requirements

The robotic kit car should have a Line Detector (for detecting the locations of each line across the track) and an Ultrasonic Sensor which can detect the distance to the feature wall. The Ultrasonic Sensor will have to be rotated so that it points towards the wall.

The robot will travel along the 1-dimensional track in a straight line. Once it reaches the end of the track, the robot can reverse back in a straight line towards the start. This optional path allows the robot to 'close the loop' on its location and increase the accuracy of its measurements to the feature wall. A robot with either turning front wheels or mecanum wheels should suit this experiment.

```python
import numpy as np
import matplotlib.pyplot as plt

ACTUAL_STARTING_POINT = -0.15 # The robot actually starts 15 cm back from the first landmark
ACTUAL_ENDING_POINT = 4.15 # The robot actually ends 15 cm past the last landmark
ACTUAL_LANDMARK_LOCATIONS = [0.00, 1.00, 2.00, 3.00, 4.00] # Painted lines are exactly 1 meter apart
ACTUAL_STARTING_VELOCITY = 0.008 # The robot actually travels at 0.8 cm per second while it believes it is traveling at 1 cm per second
EXPERIMENT_01_ACCELERATION = 0.00 # The robot does not accelerate in Experiment #1
EXPERIMENT_02_ACCELERATION = 0.0001 # The robot accelerates in Experiment #2 but continues to believe it is traveling at a constant 1 cm per second
BELIEVED_STARTING_POINT = -0.10 # The robot believes that it is starting 10 cm back from the first landmark
BELIEVED_DISTANCE_BETWEEN_WAYPOINTS = 0.01 # The robot will guess each measurement waypoint is 1 cm from the last
LOW_CONFIDENCE_WAYPOINT = 1.0 # The low confidence of the guessed waypoint added to the omega_matrix and xi_vector
HIGH_CONFIDENCE_LANDMARK = 100.0 # The high confidence of the painted landmarks added to the omega_matrix and xi_vector
```

```python
def is_landmark(actual_location, next_actual_location):
    """
    In this virtual experiment, the landmark is assumed to have zero width and is detected once
    as the robot passes over.

    In a real-world experiment, this landmark detection would need to be modified to account for the
    width of the painted line. Each landmark still must only be detected once as the robot passes over.
    As a real-world robot will not know which landmark it is passing over, the Line Detector will
    need to be sampled must faster than the waypoint measurements to guarantee that the landmark
    is not missed.
    """
    for landmark in ACTUAL_LANDMARK_LOCATIONS:
        if (actual_location <= landmark) and (next_actual_location > landmark):
            return landmark
    return None
```

```python
def generate_actual_measurements(actual_acceleration=0.00):
    """
    This routine will generate the true x-locations of the robot along the 1-dimensional track
    and the true perpendicular y-distance from the robot to the feature wall.
    This routine is not required for a real-world implementation.
    Instead, the omega_matrix and xi_vector used by Graph SLAM, and the y_distances collected
    from the feature wall, can be generated directly as the robot moves down the track. 
    """
    actual_x = ACTUAL_STARTING_POINT
    actual_velocity = ACTUAL_STARTING_VELOCITY
    actual_xy = [] # array of actual (x,y) measurements
    while actual_x <= ACTUAL_ENDING_POINT:
        actual_y = np.sin(2.0 * actual_x) + 2.0 # The feature wall is shaped like a sine-wave in this virtual experiment
        actual_xy.append([actual_x, actual_y])
        actual_x += actual_velocity
        actual_velocity += actual_acceleration
    return actual_xy
```

```python
def generate_omega_xi(actual_xy):
    """
    This routine converts the actual x-locations and y-distances into the naive measurements
    believed by the robot. These are expressed in the omega_matrix and xi_vector used by
    the Graph SLAM algorithm, as well as the believed distance measurements to the feature wall.
    """
    omega_matrix = np.zeros([len(actual_xy), len(actual_xy)], dtype=float) # Initialize the datasets based upon the size of the actual measurements
    xi_vector = np.zeros(len(actual_xy), dtype=float)
    believed_x = BELIEVED_STARTING_POINT
    believed_xy = [] # array of actual (x,y) measurements
    for i in range(len(actual_xy) - 1):
        [actual_x, actual_y] = actual_xy[i]
        [next_actual_x, next_actual_y] = actual_xy[i + 1]
        believed_xy.append([believed_x, actual_y]) # Mix the actual_y measurement with the believed_x location
        believed_x += BELIEVED_DISTANCE_BETWEEN_WAYPOINTS

        # Add the naive distances between the waypoints to the omega_matrix and xi_vector
        omega_matrix[i, i] += 1 * LOW_CONFIDENCE_WAYPOINT # distance between waypoint i and waypoint i+1
        omega_matrix[i, i+1] += -1 * LOW_CONFIDENCE_WAYPOINT
        omega_matrix[i+1, i] += -1 * LOW_CONFIDENCE_WAYPOINT
        omega_matrix[i+1, i+1] += 1 * LOW_CONFIDENCE_WAYPOINT
        xi_vector[i] += -BELIEVED_DISTANCE_BETWEEN_WAYPOINTS * LOW_CONFIDENCE_WAYPOINT # location difference between waypoints
        xi_vector[i+1] += BELIEVED_DISTANCE_BETWEEN_WAYPOINTS * LOW_CONFIDENCE_WAYPOINT

        # Add the precise landmarks to the omega_matrix and xi_vector
        landmark_location = is_landmark(actual_x, next_actual_x)
        if landmark_location:
            omega_matrix[i, i] += 1 * HIGH_CONFIDENCE_LANDMARK
            xi_vector[i] += landmark_location * HIGH_CONFIDENCE_LANDMARK

    believed_xy.append([believed_x, next_actual_y]) # Mix the actual_y measurement with the believed_x location
    return believed_xy, omega_matrix, xi_vector
```

## Experiment 1

Start the robot about 10 cm back from the first painted landmark, then drive the robot at constant speed. The robot has no idea how fast it is moving, but it will take a waypoint sample every 1 second and guess that it has traveled 1 cm since the last waypoint sample. Each is then added to the omega-matrix and xi-vector used by the Graph SLAM algorithm as a 'low confidence' sample. If the robot detects a painted line, then it will also add it to the omega-matrix and x-vector as a 'high confidence' landmark.

From the low-confidence waypoints and high-confidence landmarks, the Graph SLAM algorithm will calculate a more accurate set of x-locations as a mu-vector. Note from the plot how the robot believes that it has traveled more than 5 meters, while the Graph SLAM algorithm more accurately determines that it has only traveled a little more than 4 meters.

```python
# Robot acceleration = 0.000 meters / second
actual_xy = generate_actual_measurements(actual_acceleration=EXPERIMENT_01_ACCELERATION)
believed_xy, omega_matrix, xi_vector = generate_omega_xi(actual_xy)
```

```python
# Calculate the more accurate x_locations using the Graph SLAM algorithm to calculate mu
calculated_mu_x = np.linalg.inv(np.matrix(omega_matrix)) * np.expand_dims(xi_vector, 0).transpose()
```

```python
# Compare the original belief of the feature wall against the more accurate shape calculated from Graph SLAM
plt.title("Shape of the Feature Wall")
believed_feature_wall_zip = list(zip(*believed_xy))
plt.scatter(list(believed_feature_wall_zip[0]), list(believed_feature_wall_zip[1]), label='Believed Shape', s=5)
plt.scatter(list(calculated_mu_x), list(believed_feature_wall_zip[1]), label='Actual Shape', s=5)
plt.xlabel('x-location along 1-dimensional track')
plt.ylabel('y-measurement to feature wall') 
plt.legend(loc='lower left')
plt.show()
```

## Experiment 2

In this second experiment, the robot is accelerating as it drives along the track. Again start the robot about 10 cm back from the first painted landmark. While the robot still has no idea how fast it is moving, and does not know it is also accelerating, it will take a waypoint sample every 1 second and guess that it has traveled 1 cm since the last waypoint sample. As before, the robot will add 'low confidence' waypoints and 'high confidence' landmarks to the omega-matrix and xi-vector.

From the plot, the robot believes that it has only traveled 2 meters and that the feature wall is shaped like a compressed sine-wave. The Graph SLAM algorithm stretches the feature wall over a more accurate 4 meters.

```python
# Robot acceleration = 0.0001 meters / second
actual_xy = generate_actual_measurements(actual_acceleration=EXPERIMENT_02_ACCELERATION)
believed_xy, omega_matrix, xi_vector = generate_omega_xi(actual_xy)
```

```python
# Calculate the more accurate x_locations using the Graph SLAM algorithm to calculate mu
calculated_mu_x = np.linalg.inv(np.matrix(omega_matrix)) * np.expand_dims(xi_vector, 0).transpose()
```
